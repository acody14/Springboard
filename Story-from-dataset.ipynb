```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('social_media.csv')
```

```python
#removing final row where all values are 0
df = df.drop(index=177,axis=0)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

# Questions
## 1. What trends do we observe in the data?
## 2. What platforms previously had a large share of the market that no longer do?
## 3. What platforms previously had a small share of the market that now have a large share of the market?

```python
df_long = pd.melt(df,
                  id_vars=['Date'],
                  var_name='platform',
                  value_name='usage')

sns.lineplot(data=df_long,
             x='Date',
             y='usage', 
             hue='platform',palette='viridis')
plt.title('Global Social Media Usage by platform')
plt.xlabel('Date')
plt.ylabel('Percentage of the Market')
plt.legend(bbox_to_anchor=(1,1))
```

# Question 1: What trends do we observe in the data?
## a. Facebook's continued Dominance
## b. Quick decline in non-FB platforms between 2009 and 2012
## c. ascent in some platforms between 2017 and 2019
## d. change in market composition in 2020

```python
platforms = ['Facebook','Twitter','YouTube','Instagram','StumbleUpon']
for platform in platforms:
    plt.plot(df['Date'],df[platform])
    plt.title('Global Social Media Usage of {}'.format(platform))
    plt.xlabel('Date')
    plt.ylabel('Percentage of the Market')
    plt.show()
```
