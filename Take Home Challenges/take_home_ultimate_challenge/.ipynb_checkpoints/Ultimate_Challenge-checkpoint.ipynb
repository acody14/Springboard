```python
df = pd.read_json('ultimate_data_challenge.json')
df.sort_values('signup_date',ascending=True).head()
```

```python
df.info()
```

```python
# date columns are strings right now

df['signup_date'] = pd.to_datetime(df['signup_date'])
df['last_trip_date'] = pd.to_datetime(df['last_trip_date'])
```

```python
# checking for missing values
df.isna().sum()
```

```python
# imputing/dropping missing data
df.dropna(subset='phone',axis=0, inplace=True) # we can't fill in phone information that isn't there
df['avg_rating_of_driver'] = df['avg_rating_of_driver'].fillna(df['avg_rating_of_driver'].median()) # filling others with median value
df['avg_rating_by_driver'] = df['avg_rating_by_driver'].fillna(df['avg_rating_by_driver'].median()) 
```

```python
df.isna().sum().any()
```

```python
# Creating Active Column to be our target variable

#Current date not listed in Instructions so we'll assume it's the last date listed in the data
current_date = df['last_trip_date'].max()
print(current_date)

# Creating Active Column
df['active'] = df['last_trip_date'].apply(lambda last_trip_date: 1 if (current_date - last_trip_date).days <= 30 else 0)
```

```python
# visualizing active vs inactive users
plt.figure(figsize=(3,5))
counts = df['active'].value_counts()
plt.bar(counts.index, counts.values, width=0.3)
plt.xticks([0,1],['inactive','active'])
plt.ylabel('Count')
plt.show()

print('number of active users:   ',counts[1])
print('number of inactive users: ',counts[0])

assert counts[0] + counts[1] == len(df)
percent_active = round(counts[1] / len(df) * 100, 2)
print(f'percent of users who are active: {percent_active}%')
```

```python
# our categories are imbalanced - we might need to use stratified k fold cross validation to preserve class proportions in our training data
```

```python
df.head()
```

```python
# Let's examine more relationships between variables and our target active
numerical_vars = ['trips_in_first_30_days','avg_rating_of_driver','avg_surge','surge_pct','weekday_pct','avg_dist','avg_rating_by_driver']
for column in numerical_vars:    
    sns.violinplot(data=df,
                   x='active',
                   y=column)
    plt.title(column + ' by active status')
    plt.xticks([0,1],['inactive','active'])
    plt.ylabel(column)
    plt.show()
    
```

```python
# not too much observable difference between these.
```

```python
#Creating dummy variables from categorical columns

cat_columns = ['city','phone']
dummies = pd.get_dummies(df[cat_columns])
df_dummies = pd.concat([df,dummies],axis=1)
df_dummies.drop(cat_columns,axis=1, inplace=True)
```

```python
# dropping non-feature variables - we won't predict using the date columns
df_dummies.drop(['signup_date','last_trip_date'],axis=1,inplace=True)

# dropping target variable as well
df_dummies.drop('active',axis=1,inplace=True)
```

## Model Building

```python
#splitting our data into train and test sets

X = df_dummies.values
y = df['active']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

```python
# scaling data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# K Nearest Neighbors Classifier with 6 for k
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train_scaled,y_train)
knn.score(X_test_scaled,y_test)
```

```python
train_accuracies = {}
test_accuracies = {}

neighbors = np.arange(1,25) # range opf our k values
for neighbor in neighbors:

    #Instantiating new model with new number of neighbors
    knn = KNeighborsClassifier(n_neighbors=neighbor)

    # using cross validation to train and score the new model (we don't need to fit the model first)
    skf = StratifiedKFold(n_splits=10, shuffle=True)
    cv_results_train = cross_val_score(knn,X_train_scaled,y_train,cv=skf)
    cv_results_test = cross_val_score(knn,X_test_scaled,y_test,cv=skf)
    
    train_accuracies[neighbor] = np.mean(cv_results_train)
    test_accuracies[neighbor] = np.mean(cv_results_test)
    
plt.figure(figsize=(8,6))
plt.title('KNN: Varying Number of Neighbors')
plt.plot(neighbors, train_accuracies.values(), label='Training Accuracy')
plt.plot(neighbors, test_accuracies.values(), label='Testing Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
    
```

```python
#Looks like testing accuracy maximizes around 15 neighbors:
best_knn = KNeighborsClassifier(n_neighbors=15)
skf = StratifiedKFold(n_splits=10, shuffle=True)
cv_results_train = cross_val_score(knn,X_train_scaled,y_train,cv=skf)
cv_results_test = cross_val_score(knn,X_test_scaled,y_test,cv=skf)

print('Best Model: 15 neighbors')
print('Train accuracy:',np.mean(cv_results_train))
print('Test accuracy:',np.mean(cv_results_test))
```

```python
# To predict we still have to fit an individual model even though we didn't to use cross validation
best_knn.fit(X_train_scaled,y_train)
y_pred = best_knn.predict(X_test_scaled)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred, target_names=['inactive','active']))
```

```python
# This model is a decent predictive model HOWEVER it's not helpful when it comes to identifying important features
# So I will evaluate another random forest classifier
```

## More Useful Model - Random Forest Classifier

```python
# I'll use Bayesian Optimization for Hyper Parameter Tuning - adapted the code from my capstone 2 modeling notebook
from sklearn.ensemble import RandomForestClassifier
from bayes_opt import BayesianOptimization

#optimizes accuracy for our random forest classifier
def rfr_objective_function(n_estimators,max_depth,min_samples_split,min_samples_leaf,max_features,max_leaf_nodes):
    rfr = RandomForestClassifier(n_estimators=int(n_estimators),
                                max_depth=int(max_depth),
                                min_samples_split=int(min_samples_split),
                                min_samples_leaf=int(min_samples_leaf),
                                max_features=int(max_features),
                                max_leaf_nodes=int(max_leaf_nodes))

    # cross validating inside our objective functions avoids overfitting to our individual set of data
    kf = StratifiedKFold(n_splits=10, shuffle=True)
    cv_scores = cross_val_score(rfr, X_train_scaled, y_train, cv=kf)

    return np.mean(cv_scores)

rfr_bo = BayesianOptimization(f = rfr_objective_function,
                              pbounds = {'n_estimators':(10,500),
                                           'max_depth':(3,10),
                                           'min_samples_split':(2,10),
                                           'min_samples_leaf':(1,10),
                                           'max_features':(3,len(df_dummies.columns)),
                                           'max_leaf_nodes':(2,30)})

rfr_bo.maximize(n_iter=15, init_points=5)
```

```python
print("Optimal Parameters for Random Forest Classifier")
for key, value in rfr_bo.max['params'].items():
    print("{}: {}".format(key,value))
```

```python
rfr_optimized = RandomForestClassifier(max_depth = int(rfr_bo.max['params']['max_depth']),
                                      max_features = int(rfr_bo.max['params']['max_features']),
                                      max_leaf_nodes = int(rfr_bo.max['params']['max_leaf_nodes']),
                                      min_samples_leaf = int(rfr_bo.max['params']['min_samples_leaf']),
                                      min_samples_split = int(rfr_bo.max['params']['min_samples_split']),
                                      n_estimators = int(rfr_bo.max['params']['n_estimators']))
rfr_optimized.fit(X_train_scaled,y_train)
y_pred_test = rfr_optimized.predict(X_test_scaled)
y_pred_train = rfr_optimized.predict(X_train_scaled)

# Evaluating Best Random Forest Classifier
print("Bayesian Optimized Random Forest Classifier:\n")
print('Train Set:')
print(confusion_matrix(y_train,y_pred_train))
print(classification_report(y_train,y_pred_train))
print('\nTest Set:')
print(confusion_matrix(y_test,y_pred_test))
print(classification_report(y_test,y_pred_test))
```

```python
# This model is a lot better than the previous one anyway
```

```python
feature_importances = rfr_optimized.feature_importances_
df_important = pd.DataFrame(zip(df_dummies.columns, feature_importances), columns=['feature','importance'])
df_important.sort_values('importance',ascending=False)
```
